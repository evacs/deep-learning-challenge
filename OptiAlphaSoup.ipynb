## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
columns_to_drop = ['EIN','NAME']
clean_app_df = application_df.drop(columns_to_drop, axis=1)
```

```python
# Determine the number of unique values in each column.
unique_counts = clean_app_df.nunique()
unique_counts
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = clean_app_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 528
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = clean_app_df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
cutoff = 1
cl_counts_over1 = class_counts[class_counts > cutoff].index
cl_counts_over1.value_counts()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = class_counts[class_counts < 1883].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
encoded_df = pd.get_dummies(clean_app_df)
encoded_df.head()
```

```python
encoded_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = encoded_df["IS_SUCCESSFUL"].values
X = encoded_df.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  81
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer    relu   sigmoid  softmax
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Check the structure of the model
nn.summary()
```

```python
nn1 = tf.keras.models.Sequential()

nn1.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))
nn1.add(tf.keras.layers.BatchNormalization())
nn1.add(tf.keras.layers.Dropout(0.5))

nn1.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))
nn1.add(tf.keras.layers.BatchNormalization())
nn1.add(tf.keras.layers.Dropout(0.5))

nn1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Compile the model with an adjusted learning rate
nn1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="binary_crossentropy", metrics=["accuracy"])
```

```python
from keras.callbacks import Callback

# Custom callback to print information every n epochs
class PrintEveryNEpochs(Callback):
    def __init__(self, n=5):
        self.n = n

    def on_epoch_end(self, epoch, logs=None):
        if (epoch + 1) % self.n == 0:
            print(f'Epoch {epoch + 1}: {logs}')
```

```python
# Train the model
print_callback = PrintEveryNEpochs(n=5)

fit_model = nn1.fit(X_train_scaled,y_train,epochs=100,verbose = 0,callbacks=[print_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Train the model
nn2 = Sequential([
    Dense(108, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation = 'sigmoid')
])

nn2.compile(optimizer='adam', loss='binary_crossentropy', metrics=["accuracy"],)

# Create an instance of the custom callback
print_callback = PrintEveryNEpochs(n=5)

nn2.fit(X_train_scaled, y_train, epochs=100, batch_size=32, verbose=0, callbacks=[print_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn2.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Train the model
nn3 = Sequential([
    Dense(108, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation = 'sigmoid')
])

nn3.compile(optimizer='adam', loss='binary_crossentropy', metrics=["accuracy"],)

# Create an instance of the custom callback
print_callback = PrintEveryNEpochs(n=5)

nn3.fit(X_train_scaled, y_train, epochs=100, batch_size=16, verbose=0, callbacks=[print_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
